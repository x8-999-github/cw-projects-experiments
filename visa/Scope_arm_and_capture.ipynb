# Experiment to connect to the Rigol MSO1104Z scope

The web is full of hacks to connect to a rigol scope, Sometimes with cutsomized libraries, sometimes by having to install visa driver and such. The code here is based on example code found in DSRemote and DS1054Z_screen_capture and tries to be a bit neater in the hope to get it into the CW repositories https://github.com/python-ivi/python-vxi11

```python
import vxi11
instr =  vxi11.Instrument("10.16.82.186")
help(instr)
```

```python
instr.ask("*IDN?")
```

```python
# Get the screen 
from PIL import Image
import codecs
import io

instr.write(":DISP:DATA?")
data = instr.read(encoding ="latin1")

#
# TMC header starts with # , the number of digits and the data itself
# in this case it is a bmp image 
digit_count = int(data[1])
data_size = int(data[2:2+digit_count])
img_data = data[2+digit_count:2+digit_count +data_size]

Image.open(io.BytesIO(img_data.encode("latin1")))
```

# ARM and get the data out

Get traces.

```python
import numpy
import matplotlib.pyplot as plot
import time
%matplotlib inline


timeout_s = 20

# ARM

instr.write(":SINGLE")
stat = instr.ask(":TRIG:STAT?")
time_end = time.time() + timeout_s

while time_end > time.time():
    stat = instr.ask(":TRIG:STAT?")
    print(stat)
    if stat == "STOP":
        break
    #break if stat == "WAIT":
    time.sleep(0.2)

stat = instr.ask(":TRIG:STAT?")

if stat == "STOP":
    for i in  ("CHAN1","CHAN2","CHAN3","CHAN4"):
        instr.write(":WAV:SOUR %s" % i)
        instr.write(":WAV:DATA?")
        rawdata = instr.read(encoding="latin1")
        mydata = numpy.frombuffer(rawdata, dtype='B', offset=10)
        plot.plot(mydata)
    plot.show()
```

```python
# Grab the data from channel 1
instr.write(":STOP")
instr.write(":WAV:MODE NORM")
instr.write(":WAV:MODE STAR 0")
instr.write(":WAV:FORM BYTE")


for i in  ("CHAN1","CHAN2","CHAN3","CHAN4"):
    instr.write(":WAV:SOUR %s" % i)
    instr.write(":WAV:DATA?")
    rawdata = instr.read(encoding="latin1")
    mydata = numpy.frombuffer(rawdata, dtype='B', offset=10)
    plot.plot(mydata)


 
plot.show()
```
